```python
import os
import pandas as pd
import re
import numpy as np
from pathlib import Path
from scipy.stats import bootstrap
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression, LogisticRegressionCV
from sklearn.metrics import classification_report,accuracy_score, confusion_matrix, roc_curve, auc
import statsmodels.api as sm
from scipy.stats import norm
from sklearn.model_selection import StratifiedKFold, cross_val_score
from sklearn.feature_selection import RFE, SelectKBest, f_classif
from joblib import Parallel, delayed
from statsmodels.stats.outliers_influence import variance_inflation_factor
from sklearn.utils import resample
from sklearn.preprocessing import StandardScaler, LabelEncoder, label_binarize
from sklearn.naive_bayes import GaussianNB, MultinomialNB

def natural_sort_key(s):
    return [int(text) if text.isdigit() else text for text in re.split(r'(\d+)', s)]

source_dir = "../data/AReM" 
target_dir = "../data/Splitted datasets"  
os.makedirs(os.path.join(target_dir, "train"), exist_ok=True)
os.makedirs(os.path.join(target_dir, "test"), exist_ok=True)

for activity in ['bending1','bending2','cycling','lying','sitting','standing','walking']:
    activity_path = os.path.join(source_dir, activity)
    files = sorted([f for f in os.listdir(activity_path)], key=natural_sort_key)
    test_files = files[:3] if "bending" not in activity else files[:2]
    train_files = files[len(test_files):]
    for file in test_files:
        df = pd.read_csv(os.path.join(activity_path, file), skiprows=4, usecols=range(7)) 
        df.to_csv(os.path.join(target_dir, "test", f"{activity}_{file}"), index=False)
        
    for file in train_files:
        df = pd.read_csv(os.path.join(activity_path, file), skiprows=4, usecols=range(7))
        df.to_csv(os.path.join(target_dir, "train", f"{activity}_{file}"), index=False)
print('Train and test set are stored in folder splitted datasets')
```

1.(c) feature extraction i. ii.

```python
datasets_dir = "../data/Splitted datasets"  
feature_dir = "../data/Feature extracted"
os.makedirs(feature_dir, exist_ok=True)

def extract_features(file_path):
    df = pd.read_csv(file_path, usecols=range(7))
    features = {}
    for col in df.columns:
        features[f"{col}_min"] = df[col].min()
        features[f"{col}_max"] = df[col].max()
        features[f"{col}_mean"] = df[col].mean()
        features[f"{col}_median"] = df[col].median()
        features[f"{col}_std"] = df[col].std()
        features[f"{col}_q1"] = df[col].quantile(0.25)
        features[f"{col}_q3"] = df[col].quantile(0.75)
    feature_df = pd.DataFrame([features])
    feature_df.insert(0, "instance", os.path.basename(file_path)) 
    return feature_df

for dataset in ["train", "test"]:
    feature_list = []
    dataset_path = os.path.join(datasets_dir, dataset)
    for file in os.listdir(dataset_path):
        if file == ".ipynb_checkpoints":   #ignore checkpoint folder
            continue
        file_path = os.path.join(dataset_path, file)
        if file == 'bending2_dataset4.csv':    #cleaning bending2_dataset4, which has a wrong form
            with open(file_path,"r",encoding = 'utf-8') as f:
                lines = f.readlines()
            title = lines[0].strip()
            datas = [line.strip().replace(" ", ",") for line in lines[1:]]
            with open(file_path, "w", encoding="utf-8") as f:
                f.write(title + "\n") 
                f.write("\n".join(datas)) 
        feature_list.append(extract_features(file_path))
    all_features = pd.concat(feature_list, ignore_index=True)
    all_features.to_csv(os.path.join(feature_dir, f"{dataset}_features.csv"), index=False)
print('all features are stored in test_features.csv and train_features.csv under folder Features extracted')
```

1.(c) feature extraction iii. iv.

```python
def compute_bootstrap_ci(df, confidence_level=0.90):
    ci_results = {}
    for col in df.columns[1:]:  
        data = df[col].values
        std_dev = np.std(data, ddof=1) 
        res = bootstrap((data,), np.std, confidence_level=confidence_level, n_resamples=1000, method='percentile')
        ci_results[col] = {
            "std": std_dev,
            "ci_lower": res.confidence_interval.low,
            "ci_upper": res.confidence_interval.high
        }
    return pd.DataFrame(ci_results).T

train_features = pd.read_csv(os.path.join(feature_dir, "train_features.csv"))
test_features = pd.read_csv(os.path.join(feature_dir, "test_features.csv"))
train_ci = compute_bootstrap_ci(train_features)
test_ci = compute_bootstrap_ci(test_features)
train_ci.to_csv(os.path.join(feature_dir, "train_std_ci.csv"))
test_ci.to_csv(os.path.join(feature_dir, "test_std_ci.csv"))

# choosing features(min, mean, max)
important_features = ["min", "mean", "max"]
selected_features = [col for col in train_features.columns if any(feature in col for feature in important_features)]
train_selected = train_features[["instance"] + selected_features]
test_selected = test_features[["instance"] + selected_features]
train_selected.to_csv(os.path.join(feature_dir, "train_selected_features.csv"), index=False)
test_selected.to_csv(os.path.join(feature_dir, "test_selected_features.csv"), index=False)

print("std and bootstrap ci stored in folder feature extracted")
```

Next we use l in range(1,21) to split the time series

```python
def extract_features_new(time_series, column_name, segment_idx):
    features = {}
    features[f'{column_name}_min_{segment_idx}'] = np.min(time_series)
    features[f'{column_name}_max_{segment_idx}'] = np.max(time_series)
    features[f'{column_name}_mean_{segment_idx}'] = np.mean(time_series)
    features[f'{column_name}_median_{segment_idx}'] = np.median(time_series)
    features[f'{column_name}_std_{segment_idx}'] = np.std(time_series)
    features[f'{column_name}_q1_{segment_idx}'] = np.percentile(time_series, 25)
    features[f'{column_name}_q3_{segment_idx}'] = np.percentile(time_series, 75)
    return features

def process_file(file_path, l):
    df = pd.read_csv(file_path, usecols=range(7))  
    num_time_steps = df.shape[0]  
    segment_size = num_time_steps // l  

    all_features = []

    for i in range(l):
        instance_features = []  
        for col in df.columns:
            start_idx = i * segment_size
            end_idx = (i + 1) * segment_size if i < l - 1 else num_time_steps  
            segment = df[col].iloc[start_idx:end_idx]  
            features = extract_features_new(segment.values,col,i+1)  
            instance_features.extend(features.values()) 
        
        all_features.extend(instance_features) 
    return pd.DataFrame(all_features)

def modify_column_names(df):
    num_columns = len(df.columns)
    feature_types = ['min', 'max', 'mean', 'median', 'std', 'q1', 'q3']
    base_names = ['time', 'avg_rss12', 'var_rss12', 'avg_rss13', 'var_rss13', 'avg_rss23', 'var_rss23']
    new_columns = ['instance']
    group_count = 1
    current_base_idx = 0
    for i in range(1, num_columns): 
        base_name = base_names[current_base_idx]
        feature_type = feature_types[(i - 1) % len(feature_types)] 
        new_columns.append(f'{base_name}_{feature_type}_{group_count}')
        if (i) % 7 == 0:
            current_base_idx += 1
            if current_base_idx >= len(base_names):
                current_base_idx = 0
                group_count += 1
    df.columns = new_columns
    return df

input_folder = '../data/Splitted datasets/train'  
output_folder = '../data/logistic features'  
files = os.listdir(input_folder)
os.makedirs(output_folder, exist_ok=True)

for l in range(1, 21):  
    all_instances_features = []
    for file in files:
        if file.endswith('.csv'): 
            file_path = os.path.join(input_folder, file)
            features_df = process_file(file_path, l).T
            features_df.insert(0, 'instance', file)  
            all_instances_features.append(features_df) 
    result_df = pd.concat(all_instances_features, ignore_index=True)
    result_df = modify_column_names(result_df)
    output_file = os.path.join(output_folder, f'features_l_{l}.csv')
    result_df.to_csv(output_file, index=False)
    print(f"Saved {output_file}")
```

then we start logistic regression with these files, when first calculating p values, I can't get the exact value, even if I set the max_iter parameter and delete the constant columns, since the variables are highly collinear. I detailed how to calculate that in the #notes.

```python
np.seterr(divide='ignore', invalid='ignore') #ignore errors in calculating pvalue

def load_data(l, input_folder):
    file_path = os.path.join(input_folder, f'features_l_{l}.csv')
    df = pd.read_csv(file_path)
    df['activity'] = df['instance'].apply(lambda x: 1 if 'bending' in x else 0)
    X = df.drop(columns=['instance', 'activity'])  
    X = X.loc[:, ~X.columns.str.contains('time')] #delete time info
    X = X.loc[:, (X.nunique() > 1)]  #delete constant columns
    y = df['activity']  
    return X, y

#since we can't use sklearn to know the p_value, and if I don't filter the features at this step, statsmodel keep reporting
#errors, so I have to manually calculate P value as footnote suggested. I asked ChatGPT how to calculate pvalue.
#As we can see, these p values are close to 1, which is meaningless. Please note, I replaced NaN with 0, 0 has no meaning
#in this result.
def compute_p_values(X, y):
    model = LogisticRegression(max_iter=50, solver='liblinear')  # using sklearn and set max_iter
    model.fit(X, y)
    coef = model.coef_.flatten() 
    # è®¡ç®— Hessian çŸ©é˜µçš„è¿‘ä¼¼é€†çŸ©é˜µ
    X_with_const = np.hstack([np.ones((X.shape[0], 1)), X])  # æ·»åŠ å¸¸æ•°é¡¹
    prob = model.predict_proba(X)[:, 1]  # é¢„æµ‹æ¦‚çŽ‡
    W = np.diag(prob * (1 - prob))  # æƒé‡çŸ©é˜µ
    Hessian = X_with_const.T @ W @ X_with_const  # è®¡ç®— Hessian çŸ©é˜µ
    cov_matrix = np.linalg.inv(Hessian)  # è®¡ç®—åæ–¹å·®çŸ©é˜µï¼ˆHessian çŸ©é˜µçš„é€†ï¼‰
    # è®¡ç®—æ ‡å‡†è¯¯å·®
    se = np.sqrt(np.maximum(0, np.diag(cov_matrix)))
    # è®¡ç®— Z ç»Ÿè®¡é‡
    Z_scores = coef / se[1:]
    # è®¡ç®— p-value
    p_values = 2 * (1 - norm.cdf(np.abs(Z_scores)))
    return p_values, coef

# éåŽ†ä¸åŒçš„ l å€¼å¹¶è®¡ç®— p-value
input_folder = '../data/logistic features'
for l in range(1, 21):
    X, y = load_data(l, input_folder)
    try:
        p_values, coef = compute_p_values(X, y)
        print(f"Results for l = {l}:")
        print(f"P-Values: {p_values}")
    except np.linalg.LinAlgError:
        print(f"Singular matrix error at l = {l}, skipping...")
    print("="*50)
```

Since we can't filter features from pvalue, let's try backward selection part using RFE and 5-fold cross validation. The right way is to optimize (l,p) at the same time rather than optimize p and using this p for different l.

```python
input_folder = '../data/logistic features'

best_score = -np.inf
best_params = {'l': None, 'p': None}
best_features = []

file_data = {}
for l in range(1, 21):
    file_path = os.path.join(input_folder, f'features_l_{l}.csv')
    file_data[l] = pd.read_csv(file_path)
    file_data[l]['activity'] = file_data[l]['instance'].apply(lambda x: 1 if 'bending' in x else 0)
    
def evaluate_model(l, p, df):
    X = df.drop(columns=['instance', 'activity'])  
    X = X.loc[:, ~X.columns.str.contains('time')] #delete time info
    X = X.loc[:, (X.nunique() > 1)]  #delete constant columns
    y = df['activity']

    selector = SelectKBest(f_classif, k=min(p * 2, X.shape[1])) #feature pre_selection to improve calculating speed
    X_selected = selector.fit_transform(X, y)
    selected_kbest_indices = selector.get_support(indices=True)
    selected_kbest_features = X.columns[selected_kbest_indices]  #save feature names
    
    model = LogisticRegression(max_iter=500, solver='liblinear')
    rfe = RFE(model, n_features_to_select=p)
    X_rfe = rfe.fit_transform(X_selected, y)
    
    selected_rfe_features = selected_kbest_features[rfe.support_]
    # calculating cv_score
    skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42) #both shuffle and not shuffle is ok
    scores = cross_val_score(model, X_rfe, y, cv=skf, scoring='accuracy', n_jobs=-1)
    return l, p, np.mean(scores), selected_rfe_features.tolist()


# using parallel to improve calculating speed, I asked ChatGPT about this method
results = Parallel(n_jobs=-1, backend='loky')(
    delayed(evaluate_model)(l, p, df)
    for l, df in file_data.items()
    for p in range(1,20) 
    #although we have a lot of P, but if we have to many features, the model is too complicated
    #so I limited p in range(1,20) to speed up
)


for l, p, score, selected_features in results:
    if score > best_score:
        best_score = score
        best_params = {'l': l, 'p': p}
        best_features = selected_features 

print(f'Highest accuracy: {best_score:.2f}')
print(f'best (l,p) pair: {best_params}')
print(f'Features under best(l,p): {best_features}')
```

```python
input_folder = '../data/logistic features'  
test_folder = '../data/logistic test'

file_data = {}
for l in range(1, 21):
    file_path = os.path.join(input_folder, f'features_l_{l}.csv')
    file_data[l] = pd.read_csv(file_path)
    file_data[l]['activity'] = file_data[l]['instance'].apply(lambda x: 1 if 'bending' in x else 0)
    
def evaluate_model(l, p, df):
    X = df.drop(columns=['instance', 'activity'])
    X = X.loc[:, ~X.columns.str.contains('time')]  
    X = X.loc[:, (X.nunique() > 1)] 
    y = df['activity']

    k_best = max(2, min(p * 2, X.shape[1]))
    selector = SelectKBest(f_classif, k=k_best)
    X_selected = selector.fit_transform(X, y)
    selected_kbest_features = X.columns[selector.get_support(indices=True)]

    df_majority = df[df['activity'] == 0]
    df_minority = df[df['activity'] == 1]

    df_majority_downsampled = resample(df_majority, 
                                       replace=False,  
                                       n_samples=len(df_minority),  
                                       random_state=42)

    df_balanced = pd.concat([df_majority_downsampled, df_minority])
    
    X_balanced = df_balanced[selected_kbest_features]
    y_balanced = df_balanced['activity']

    model = LogisticRegression(max_iter=500, solver='liblinear')
    p_rfe = min(p, X_balanced.shape[1])
    rfe = RFE(model, n_features_to_select=p_rfe)
    X_rfe = rfe.fit_transform(X_balanced, y_balanced)
    selected_rfe_features = selected_kbest_features[rfe.support_]

    skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
    scores = cross_val_score(model, X_rfe, y_balanced, cv=skf, scoring='accuracy', n_jobs=-1)

    return l, p, np.mean(scores), selected_rfe_features.tolist()


results = Parallel(n_jobs=-1, backend='loky')(
    delayed(evaluate_model)(l, p, df)
    for l, df in file_data.items()
    for p in range(1,20) if df is not None and not df.empty
)

best_score = -np.inf
best_params = None
best_features = None

for l, p, score, selected_features in results:
    if score > best_score:
        best_score = score
        best_params = {'l': l, 'p': p}
        best_features = selected_features

print(f'Best accuracy: {best_score:.4f}')
print(f'best L,P: {best_params}')
print(f'best features: {best_features}')

```

```python
input_folder = '../data/logistic features'  
test_folder = '../data/logistic test'

file_data = {}
for l in range(1, 21):
    file_path = os.path.join(input_folder, f'features_l_{l}.csv')
    file_data[l] = pd.read_csv(file_path)
    file_data[l]['activity'] = file_data[l]['instance'].apply(lambda x: 1 if 'bending' in x else 0)

def evaluate_model_l1(l, df):

    X = df.drop(columns=['instance', 'activity'], errors='ignore')
    X = X.loc[:, ~X.columns.str.contains('time')]  
    X = X.loc[:, (X.nunique() > 1)]  
    y = df['activity']

    # We also use Case-Control Sampling here
    df_majority = df[df['activity'] == 0]
    df_minority = df[df['activity'] == 1]

    df_majority_downsampled = resample(df_majority, 
                                       replace=False,  
                                       n_samples=len(df_minority),  
                                       random_state=42)

    df_balanced = pd.concat([df_majority_downsampled, df_minority])
    X_balanced = df_balanced[X.columns]
    y_balanced = df_balanced['activity']

    scaler = StandardScaler() #as suggested in footnote8, we need to do standard scaler.
    X_balanced = scaler.fit_transform(X_balanced) 

    # L1 Logistic Regression
    model = LogisticRegressionCV(
        cv=5, 
        penalty='l1', 
        solver='liblinear', 
        scoring='accuracy', 
        max_iter=500
    )
    model.fit(X_balanced, y_balanced)

    best_C = model.C_[0]
    selected_features = X.columns[np.abs(model.coef_).sum(axis=0) > 1e-4]
    return l, best_C, np.mean(model.scores_[1]), selected_features.tolist(), scaler

# å¹¶è¡Œè®¡ç®— (l, C) ç»„åˆ
results = Parallel(n_jobs=-1, backend='loky')(
    delayed(evaluate_model_l1)(l, df)
    for l, df in file_data.items() if df is not None and not df.empty
)

best_score = -np.inf
best_params = None
best_features = None
best_scaler = None

for l, C, score, selected_features, scaler in results:
    if score > best_score:
        best_score = score
        best_params = {'l': l, 'C': C}
        best_features = selected_features
        best_scaler = scaler  

print(f'Best Acc: {best_score:.4f}')
print(f'Best (l,C) pair: {best_params}')
print(f'Best features: {best_features}')
print("this performs better and easier to implement")
```

2.(c).i. multinominal regression

```python
input_folder = '../data/logistic features'  
test_folder = '../data/logistic test'

def categorize_activity(instance):
  
    if 'bending' in instance:
        return 'bending'
    elif 'cycling' in instance:
        return 'cycling'
    elif 'lying' in instance:
        return 'lying'
    elif 'sitting' in instance:
        return 'sitting'
    elif 'standing' in instance:
        return 'standing'
    elif 'walking' in instance:
        return 'walking'
    else:
        return instance  

file_data = {}
for l in range(1, 21):
    file_path = os.path.join(input_folder, f'features_l_{l}.csv')
    file_data[l] = pd.read_csv(file_path)
    file_data[l]['activity'] = file_data[l]['instance'].apply(categorize_activity)
    
def evaluate_multiclass_l1(l, df):
    X = df.drop(columns=['instance', 'activity'])
    X = X.loc[:, ~X.columns.str.contains('time')]  
    X = X.loc[:, (X.nunique() > 1)] 
    y = df['activity']
    
    label_encoder = LabelEncoder()
    y = label_encoder.fit_transform(y)
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)  
    model = LogisticRegressionCV(
        cv=5, 
        penalty='l1', 
        solver='saga', 
        multi_class='multinomial', 
        scoring='accuracy', 
        max_iter=5000
    )
    
    model.fit(X_scaled, y)
    best_C = model.C_[0]
    selected_features = X.columns[np.abs(model.coef_).sum(axis=0) > 1e-4]
    
    return l, best_C, np.mean(model.scores_[1]), selected_features.tolist()


results = Parallel(n_jobs=-1, backend='loky')(
    delayed(evaluate_multiclass_l1)(l, df)
    for l, df in file_data.items() if df is not None and not df.empty
)


best_score = -np.inf
best_params = None
best_features = None


for l, C, score, selected_features in results:
    if score > best_score: 
        best_score = score
        best_params = {'l': l, 'C': C}
        best_features = selected_features
        
print(f'Best Acc: {best_score:.4f}')
print(f'Best l,c pair: {best_params}')
print(f'Best features: {best_features}')

#because of the numbers of features used in model is not the same, we need to retrain the best model and cannot
#directly use the former model
def retrain_best_model(df, best_features, best_C):
    X_train_selected = df[best_features]  
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train_selected)  
    y = df['activity']
    label_encoder = LabelEncoder()
    y = label_encoder.fit_transform(y)
    new_model = LogisticRegressionCV(
        cv=5,
        penalty='l1',
        solver='saga',
        multi_class='multinomial',
        scoring='accuracy',
        max_iter=5000,
        Cs=[best_C]  
    )
    new_model.fit(X_train_scaled, y)
    return new_model

best_model = retrain_best_model(file_data[best_params['l']], best_features, best_params["C"])

# test set
test_path = os.path.join(test_folder, f'features_l_{best_params["l"]}.csv')
test_df = pd.read_csv(test_path)
test_df['activity'] = test_df['instance'].apply(categorize_activity)
X_test = test_df.drop(columns=['instance', 'activity'])
X_test = X_test.reindex(columns=best_features)
y_test = test_df['activity']

scaler = StandardScaler()
X_test_scaled = scaler.fit_transform(X_test)
label_encoder = LabelEncoder()
y_test_encoded = label_encoder.fit_transform(y_test)

y_pred = best_model.predict(X_test_scaled)

test_accuracy = accuracy_score(y_test_encoded, y_pred)
test_error = 1 - test_accuracy
print(f"Accuracy on testset: {test_accuracy:.4f}")
print(f"Test error: {test_error:.4f}")

conf_matrix = confusion_matrix(y_test_encoded, y_pred)
print("\nConf matrix:")
print(conf_matrix)

print("\nClassification report:")
print(classification_report(y_test_encoded, y_pred, target_names=label_encoder.classes_))

# One-vs-Rest calculating ROC
y_test_bin = label_binarize(y_test_encoded, classes=np.arange(len(label_encoder.classes_)))
y_score = best_model.decision_function(X_test_scaled)

plt.figure(figsize=(10, 6))
for i, label in enumerate(label_encoder.classes_):
    fpr, tpr, _ = roc_curve(y_test_bin[:, i], y_score[:, i])
    roc_auc = auc(fpr, tpr)
    plt.plot(fpr, tpr, label=f'{label} (AUC = {roc_auc:.2f})')

plt.plot([0, 1], [0, 1], 'k--') 
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Multinominal ROC')
plt.legend()
plt.show()
```

2.(c)ii. Finally we try bayes classifier

```python
input_folder = '../data/logistic features'
test_folder = '../data/logistic test'

def categorize_activity(instance):
    if 'bending' in instance:
        return 'bending'
    elif 'cycling' in instance:
        return 'cycling'
    elif 'lying' in instance:
        return 'lying'
    elif 'sitting' in instance:
        return 'sitting'
    elif 'standing' in instance:
        return 'standing'
    elif 'walking' in instance:
        return 'walking'
    else:
        return instance  

file_data = {}
for l in range(1, 21):
    file_path = os.path.join(input_folder, f'features_l_{l}.csv')
    file_data[l] = pd.read_csv(file_path)
    file_data[l]['activity'] = file_data[l]['instance'].apply(categorize_activity)


def train_naive_bayes(l, df, model_type='gaussian'):
    X = df.drop(columns=['instance', 'activity'])
    X = X.loc[:, ~X.columns.str.contains('time')]  
    X = X.loc[:, (X.nunique() > 1)]  
    y = df['activity']

    label_encoder = LabelEncoder()
    y = label_encoder.fit_transform(y)
    
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X) if model_type == 'gaussian' else X  # Only gaussian need a standardscaler

    model = GaussianNB() if model_type == 'gaussian' else MultinomialNB()
    model.fit(X_scaled, y)
    
    return l, np.mean(model.predict(X_scaled) == y), model, scaler, label_encoder, X.columns.tolist()


results_gaussian = Parallel(n_jobs=-1, backend='loky')(
    delayed(train_naive_bayes)(l, df, 'gaussian')
    for l, df in file_data.items() if df is not None and not df.empty
)

results_multinomial = Parallel(n_jobs=-1, backend='loky')(
    delayed(train_naive_bayes)(l, df, 'multinomial')
    for l, df in file_data.items() if df is not None and not df.empty
)


def select_best_model(results):
    best_score = -np.inf
    best_params = None
    best_model = None
    best_scaler = None
    best_label_encoder = None
    best_features = None

    for l, score, model, scaler, label_encoder, features in results:
        if score > best_score:
            best_score = score
            best_params = {'l': l}
            best_model = model
            best_scaler = scaler
            best_label_encoder = label_encoder
            best_features = features

    return best_params, best_model, best_scaler, best_label_encoder, best_features, best_score

best_params_g, best_model_g, best_scaler_g, best_label_encoder_g, best_features_g, best_score_g = select_best_model(results_gaussian)
best_params_m, best_model_m, best_scaler_m, best_label_encoder_m, best_features_m, best_score_m = select_best_model(results_multinomial)

print(f'GaussianNB Best value for l: {best_params_g["l"]}, CV Acc: {best_score_g:.4f}')
print(f'MultinomialNB Best value for l: {best_params_m["l"]}, CV Acc: {best_score_m:.4f}')

def evaluate_model_on_test(best_params, best_model, best_scaler, best_label_encoder, best_features, model_type):
    test_path = os.path.join(test_folder, f'features_l_{best_params["l"]}.csv')
    test_df = pd.read_csv(test_path)
    test_df['activity'] = test_df['instance'].apply(categorize_activity)

    X_test = test_df.drop(columns=['instance', 'activity'])
    X_test = X_test.reindex(columns=best_features, fill_value=0)
    X_test_scaled = best_scaler.transform(X_test) if model_type == 'gaussian' else X_test  

    y_test_encoded = best_label_encoder.transform(test_df['activity'])
    y_pred = best_model.predict(X_test_scaled)

    test_accuracy = accuracy_score(y_test_encoded, y_pred)
    test_error = 1 - test_accuracy

    print(f"\n{model_type.upper()} - test Acc: {test_accuracy:.4f}")
    print(f"{model_type.upper()} - test error: {test_error:.4f}")

    conf_matrix = confusion_matrix(y_test_encoded, y_pred)
    print(f"\n{model_type.upper()} - Conf matrix:")
    print(conf_matrix)

    print(f"\n{model_type.upper()} - Report:")
    print(classification_report(y_test_encoded, y_pred, target_names=best_label_encoder.classes_))

    y_test_bin = label_binarize(y_test_encoded, classes=np.arange(len(best_label_encoder.classes_)))
    y_score = best_model.predict_proba(X_test_scaled) 

    plt.figure(figsize=(10, 6))
    for i, label in enumerate(best_label_encoder.classes_):
        fpr, tpr, _ = roc_curve(y_test_bin[:, i], y_score[:, i])
        roc_auc = auc(fpr, tpr)
        plt.plot(fpr, tpr, label=f'{label} (AUC = {roc_auc:.2f})')

    plt.plot([0, 1], [0, 1], 'k--')  
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title(f'{model_type.upper()} - ROC for naive bayes')
    plt.legend()
    plt.show()


evaluate_model_on_test(best_params_g, best_model_g, best_scaler_g, best_label_encoder_g, best_features_g, 'gaussian')
evaluate_model_on_test(best_params_m, best_model_m, best_scaler_m, best_label_encoder_m, best_features_m, 'multinomial')
```
